```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
df_train = pd.read_csv('/content/loan_train.csv')
df_test = pd.read_csv("/content/loan_test.csv")

print(df_train.head())
print(df_test.head())
```

```python
print(f"Training Data Set: \n {df_train.isnull().sum()}\n")
print(f"Testing Data Set: \n {df_test.isnull().sum()}\n")
```

```python
df_train.ffill(inplace=True)
df_test.ffill(inplace=True)

print(f"Training Data Set: \n {df_train.isnull().sum()}\n")
print(f"Testing Data Set: \n {df_test.isnull().sum()}\n")
```

```python
le = LabelEncoder()
for col in df_train.select_dtypes(include=['object']).columns:
  df_train[col] = le.fit_transform(df_train[col])

for col in df_test.select_dtypes(include=['object']).columns:
  df_test[col] = le.fit_transform(df_test[col])
```

```python
print(df_train.head())
print(df_test.head())
```

```python
X_train = df_train.drop('Loan_Status', axis =1)
X_test= df_test.copy()

y_train = df_train['Loan_Status']


print(f"{X_train.head()} \n\n  {y_train.head()}")
```

```python
model = RandomForestClassifier(n_estimators=100,random_state=42)
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
print(y_pred)
```

```python
result = pd.DataFrame({'Loan_ID': df_test['Loan_ID'],
                       'Loan_Status': y_pred})
result['Loan_Status']= result['Loan_Status'].map({1:'Yes', 0:'No'})
result.to_csv('loan_predictions.csv', index=False)
```
